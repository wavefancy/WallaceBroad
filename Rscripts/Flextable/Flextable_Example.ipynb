```r
library(flextable)
library(officer)
```

```r
data(mtcars)
head(mtcars)
```

```r
mydata = head(mtcars)
```

```r
# autofit (it adjusts widths and heights regarding to text widths and heights).
ft <- flextable(mydata)
# ft = autofit(ft)
ft
```

```r
# Customize zebra blue format.
psize = 5
ft = ft %>% theme_zebra(., odd_header = "transparent", even_header = "transparent") %>%
    hline(., border = fp_border(width = .75, color = "#007FA6",style = "dashed"), part = "body" ) %>%
    hline(., border = fp_border(width = 2, color = "#007FA6"), part = "header" ) %>%
    # line_spacing + valign not work properly in MS world, so we use padding instead.
    #line_spacing(., space=1.5, part="all") %>%
    #valign(.,valign = "bottom", part = "all") %>%
    align(.,align = "center", part = "all") %>%
    padding(.,padding.top=psize, padding.bottom = psize, part = "all") %>%
    autofit
```

```r
htmltools_value(ft)
```

```r
# Save an Flextable in MS Word file.
doc <- read_docx()
doc <- body_add_flextable(doc, value = ft)
print(doc, target = "color_blue_example.docx")
```
